```python
import numpy as np

def spherical_geodesic(iterations,lamda,del_lamda,r,phi,del_r,del_phi):
  i=0
  while(lamda<iterations):
    lamda+=del_lamda
    chris_g_phiphi=-r
    chris_g_phirphi=(1/r)
    accl_r= -chris_g_phiphi*(del_phi**2)
    accl_phi=-2*chris_g_phirphi*del_r*del_phi
    del_r+=accl_r*del_lamda
    del_phi+=accl_phi*del_lamda
    r+=del_r*del_lamda
    phi+=del_phi*del_lamda
    if i%100==0:
      deg=phi*180.0/np.pi
      print(f"Lambda values: {lamda} , Radius values: {r} ,Phi Values: {deg}")
    i+=1

if __name__=='__main__':
  iterations=1000
  lamda=0
  del_lamda=0.0001
  r=1 
  phi=0
  del_r=0
  del_phi=1
  spherical_geodesic(iterations,lamda,del_lamda,r,phi,del_r,del_phi)
```
